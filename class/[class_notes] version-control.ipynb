### Introduction to MLOps and Version Control

1. Configure Git and Github - review tutorial from required materials
2. Create a github repository - this can be done in the UI
3. Clone a github repository - this can be done in the UI + terminal (`git clone`)
4. Git basics
    - `git branch` to check local branches
    - `git status` to check files status
    - `git log` to check git logs (type `q` to exit)
    - `git add` to stage files
    - `.gitignore` file
5. Commit
    - `git commit -m '<MESSAGE>'` to commit changes
    - commit convention [here](https://gist.github.com/qoomon/5dfcdf8eec66a051ecd85625518cfd13)
        * feat: \<description>
        * feat(\<affected code>): \<description>
        * fix: \<description>
        * fix(\<affected code>): \<description>
6. Branching and merging
    - `git branch <BRANCH_NAME>` to create new branch
    - `git switch <BRANCH_NAME>` to switch to a new branch
    - `git checkout <BRANCH_NAME>` to switch to a new branch
    - `git switch -c <BRANCH_NAME>` to create and switch to the new branch
    - `git checkout -b <BRANCH_NAME>` to create and switch to the new branch
    - `git branch -d <BRANCH_NAME>` to delete branch locally
    - Branch name convention [here](https://conventional-branch.github.io/) 
7. Push and pull
    - `git pull origin <REMOTE_BRANCH>` to bring your local branch up to date with a remote branch
    - `git push origin <REMOTE_BRANCH>` to update remote branch with local changes
    - `git fetch origin` + `git checkout -b <LOCAL_BRANCH> origin/<REMOTE_BRANCH>` to download a branch from remote and store it as a local branch
8. Git commands for fixes
    - `git restore` to restore changes 
    - `git reset` to unstage changes
    - How to do rollbacks:
        * `git reset --soft HEAD~1` to undo last commit but keeps the code as staged
        * `git reset --hard HEAD~1` to rewind history by one commit and erase the changes
        * `git revert <commit-hash>` to add a new commit that undoes a previous one (commit hash is obtained with `git log`)

9. Github workflow
    - Before commiting look for changes on the remote
    - Use branches for new features
    - Open pull requests and use code reviews
    - Merge branches
